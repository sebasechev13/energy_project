```python
import pandas as pd
import numpy as np 
```

```python
# read csv's
pp_data = pd.read_csv("global_power_plant_database.csv")
cc_data = pd.read_csv("OECD_Consumption.csv")
```

```python
#find the the counties in powerplan data base 
y = pp_data["country"].unique()
#find the countries in consumption data base 
x = cc_data["LOCATION"].unique()
#find the countries in both databases.
countries_l =[]
for county in y:
    if county in x:
        countries_l.append(county)
f_country = []
for country in countries_l:
    if pp_data.loc[pp_data.country == country ,'country'].count() >2:
        f_country.append(country)
#create data frames containing only the countries in both data frames
pp_data_clean = pp_data.loc[pp_data['country'].isin(f_country)]
cc_data_clean = cc_data.loc[cc_data['LOCATION'].isin(f_country)]
#make a list of the columns we need in powerplant data base 
columnns_l = ["country","country_long","name","capacity_mw","latitude","longitude","primary_fuel","estimated_generation_gwh"]
```

```python
#select the columns needed for powerplants 
pp_data_clean1 = pp_data_clean[columnns_l]
pp_data_clean1
```

```python
cc_data_clean
```

```python
pp_data_clean
```

```python
#rename location to country to merge with pp
cc_data_clean1=cc_data_clean.rename(columns={"LOCATION": "country"})
#defining column names you want from df
col_c=["country", "Value"]
#creating and printing new df with desired columns
cc_new=cc_data_clean1[col_c]
cc_new
```

```python
#Groupby for powerplant information to calculated estimated generation gwh
pp_group=pp_data_clean1.groupby('country').sum()
pp_new=pp_group.drop(['latitude', 'longitude'], axis=1)
pp_new
```

```python
#read key csv
key = pd.read_csv("key.csv")
#merge pp_new with the key to be able to have the country name
pp_info=pd.merge(pp_new,key,how='inner',on='country')
#print
pp_info
```

```python
#merge to have capacity, generation, and consumption
pp_cc_combined= pd.merge(pp_info,cc_new,how='inner',on='country')
pp_cc_combined
```

```python
#read latlon
latlon = pd.read_csv("latlon.csv")
#change name to country_long to merge
lat_lon=latlon.rename(columns={"name": "country_long"})
col_ll=["country_long","latitude","longitude"]
latlon_clean=lat_lon[col_ll]
#merge to get lat and lon
all_data= pd.merge(pp_cc_combined,latlon_clean,how='inner',on='country_long')
#print
all_data
```

```python
#solve for toe
all_data['estimated_generation_toe']=all_data["estimated_generation_gwh"]*86/1000000
```

```python
#organizing columns and renaming to add context
cols=["country_long","country","latitude","longitude","capacity_mw","estimated_generation_toe","Value"]
all_data=all_data[cols]
all_data=all_data.rename(columns={"Value": "consumption"})

#print
all_data
```

```python
#create csv
all_data.to_csv(r'all_data.csv')
```
